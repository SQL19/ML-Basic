This project 
- create a random forest model to predict the mobile phone price range 
- use random search to find the best combination of parameters

```python
import numpy as np 
import pandas as pd
from sklearn import ensemble
from sklearn import metrics
from sklearn import model_selection
```

```python
# load data
df = pd.read_csv('mobile_train.csv')
df.head()
```

```python
# define features and label
X = df.drop('price_range', axis=1).values
y = df.price_range.values
```

```python
# random forest model
classifier = ensemble.RandomForestClassifier()
```

```python
# define a grid of parameters
param_grid = {"n_estimators": np.arange(100, 1500, 100), 
              "max_depth": np.arange(1, 31), 
              "criterion": ["gini", "entropy"]
}
```

```python
model = model_selection.RandomizedSearchCV(
    estimator=classifier, # the model we difined
    param_distributions=param_grid, # grid of parameters
    n_iter=20, # number of iterations
    scoring="accuracy", 
    verbose=10, # higher value of verbose implies a lot of details are printed
    cv=5 
)
```

```python
# fit the model and extract best score
model.fit(X, y)
print(f"Best score: {model.best_score_}")
```

```python
print("Best parameters set:")
best_parameters = model.best_estimator_.get_params() 
for param_name in sorted(param_grid.keys()):
    print(f"\t{param_name}: {best_parameters[param_name]}")
```
