```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
data = loadmat('ex6data2.mat')
X, y = data['X'], data['y']
```

```python
X.shape, y.shape
```

```python
def plot_data():
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
plot_data()
```

```python
def plot_boundary(model):
    x_min, x_max = 0, 1
    y_min, y_max = 0.4, 1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    target = np.c_[xx.flatten(), yy.flatten()]
    z = model.predict(target)
    zz = z.reshape(xx.shape)
    plt.contour(xx, yy, zz)
```

compare different gamma

```python
svc1 = SVC(C=1, kernel='rbf', gamma=1)
svc1.fit(X, y.flatten())
print(svc1.score(X, y.flatten()))
```

```python
plot_boundary(svc1)
plot_data()
plt.show()
```

```python
svc10 = SVC(C=1, kernel='rbf', gamma=10)
svc10.fit(X, y.flatten())
plot_boundary(svc10)
plot_data()
plt.show()
```

```python
svc100 = SVC(C=1, kernel='rbf', gamma=100)
svc100.fit(X, y.flatten())
plot_boundary(svc100)
plot_data()
plt.show()
```

The larger the gamma, the more complex the model, could be over-fitting if the gamma is too large.
